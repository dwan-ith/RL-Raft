```
import os
import matplotlib.pyplot as plt
from stable_baselines3 import PPO
from rlraft.env import RaftEnv
from rlraft.training import train_policy

# Train the policy with Domain Randomization (5-100 nodes)
print('Starting PPO Training with Domain Randomization...')
result = train_policy(episodes=5000, nodes=100, output_path='runs/colab_policy.json')

print(f'Training complete! Average Failover: {result.average_failover_ms:.1f}ms')
print(f'Best node win rate: {result.best_node_win_rate * 100:.1f}%')
```

```
# Visualize results
import pandas as pd
from rlraft.experiments import run_simulation_comparison

print('Running deterministic evaluation vs heuristics...')
csv_path = run_simulation_comparison(nodes=100, episodes=500, output_dir='runs')
df = pd.read_csv(csv_path)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
df.plot(kind='bar', x='policy', y='average_failover_ms', ax=ax1, color='black', legend=False)
ax1.set_title('Average Failover Time (ms)')
ax1.set_ylabel('Milliseconds')

df.plot(kind='bar', x='policy', y='best_node_win_rate', ax=ax2, color='gray', legend=False)
ax2.set_title('Best-Connected Node Win Rate')
ax2.set_ylabel('Win Rate')
plt.tight_layout()
plt.show()
display(df)
```
